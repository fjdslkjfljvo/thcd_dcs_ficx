```python
import numpy as np
import pandas as pd
from statsmodels.tsa.api import VAR

# Example data
data = {
    'Series1': [1.1, 2, 3, 4, 5],
    'Series2': [2, 3, 4, 5, 6],
    'Series3': [3, 4, 5, 6, 7]
}
df = pd.DataFrame(data)


# Fit the VAR model
model = VAR(df)
results = model.fit(maxlags=3)  # Adjust maxlags as needed

# Print model summary
print("Model Summary:")
print(results.summary())

# Forecast future values
forecast_steps = 5
forecast = results.forecast(df.values[-results.k_ar:], steps=forecast_steps)
forecast_df = pd.DataFrame(forecast, columns=df.columns)

# Display the forecasted values
print(f"\nForecast for the next {forecast_steps} periods:")
print(forecast_df)
```

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.ar_model import AutoReg
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

# Step 1: Create a synthetic time series data
np.random.seed(0)
lst = [0]
for i in range(99):
    err = np.random.randn(1)[0]
    lst.append(0.5*lst[-1] + err)
data = pd.Series( np.array(lst) )

# Step 2: Plot the time series data
plt.figure(figsize=(10, 6))
plt.plot(data)
plt.title('Synthetic Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()

# Step 3: Fit AR(2) model
model = AutoReg(data, lags=2)
model_fit = model.fit()
print(model_fit.summary())

# Step 4: Plot residuals
residuals = model_fit.resid

plt.figure(figsize=(10, 6))
plt.plot(residuals)
plt.title('Residuals of AR(2) Model')
plt.xlabel('Time')
plt.ylabel('Residuals')
plt.show()

# Plot ACF of residuals
plot_acf(residuals)
plt.show()

# Step 5: Forecast future values
n_periods = 10
predictions = model_fit.predict(start=len(data), end=len(data) + n_periods - 1)

# Plot the original data and the forecasts
plt.figure(figsize=(10, 6))
plt.plot(data, label='Original Data')
plt.plot(range(len(data), len(data) + n_periods), predictions, label='Forecasted Data', color='red')
plt.title('AR(2) Model Forecast')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
print(model_fit.params)
```
